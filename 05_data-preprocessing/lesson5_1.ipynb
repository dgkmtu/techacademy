```python
dataset2 = dataset[['球団','推定年俸','打率']]
# 列名をリネームする
dataset2 = dataset2.rename(columns={'球団': 'team', '推定年俸':'salary', '打率':'batting_avg'})
dataset2.head()
```

```python
dataset2['salary'].hist(bins=20)
```

```python
# 球団ごとの推定年俸をsalary_by_teamに格納
salary_by_team = dataset2[['team', 'salary']].groupby('team').mean().reset_index()

# 棒グラフを表示
salary_by_team.plot.bar(x='team')
```

```python
# 推定年俸上位3チームの「team, salary」を抜き出し
top3_team = dataset2.query("team in ['g', 'h', 'l']")[['team', 'salary']]

# teamごとのsalaryをboxplotで表示
ax = top3_team.boxplot(by='team')
# Boxplot ... の下のタイトル表示を抑制
ax.set_title('')
```

```python
dataset["本塁打率"] = dataset["本塁打"]/dataset["安打"]
dataset["本塁打率"].head(10)
```

```python
dataset['推定年俸の対数'] = np.log10(dataset['推定年俸'])
dataset['推定年俸の対数'].head()
```

```python
dataset3 = pd.get_dummies(data=dataset, columns=['球団'])
dataset3.head()
```

```python
# データの正規化は２種類存在する
# 正規化用にdatasetの一部を抜き出し
dataset4 = dataset[["推定年俸", "打点"]]
dataset4.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
dataset4_std = sc.fit_transform(dataset4)
dataset4_std
```

```python
# Min-Max Normalization
from sklearn.preprocessing import MinMaxScaler
ms = MinMaxScaler()
dataset4_norm = ms.fit_transform(dataset4)
dataset4_norm
```

```python
dataset.isnull().sum()
```

```python
# 欠損値に平均値を代入。「dataset['年数'].mean()」で平均値を求めている
dataset['本塁打率'] = dataset['本塁打率'].fillna(dataset['本塁打率'].mean())
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
# 欠損値を含む行を削除
dataset = dataset.dropna()
dataset.head()
```

```python
dataset.corr()
```

```python
dataset2 = dataset[['推定年俸', '打率', '出塁率']]

# 列名をリネームする
dataset2 = dataset2.rename(columns={'推定年俸': 'salary', '打率': 'batting_avg', '出塁率': 'on_base_per'})
dataset2.head()
```

```python
dataset2.plot(kind='scatter', x='batting_avg', y='on_base_per')
```

```python
# 推定年俸：打率
dataset2.plot(kind='scatter', x='salary', y='batting_avg')
```

```python
dataset2.plot(kind='scatter', x='salary', y='on_base_per')
```

```python
from sklearn.decomposition import PCA

# batting_avgとon_base_perを取り出して主成分分析
pca = PCA(n_components=1).fit(dataset2[['batting_avg', 'on_base_per']])

# 主成分分析した結果を基にデータ変換を行い取得する
values = pca.transform(dataset2[['batting_avg', 'on_base_per']])

# 得られた成分をdataset2に追加
dataset2['PCA_val'] = values[:,0]
```

```python
# 推定年俸：PCA分析
dataset2.plot(kind='scatter', x='salary', y='PCA_val')
```
